# Faculty Data Export Utility

This notebook provides a simple way to export the entire faculty database (109 records) into CSV and JSON formats.

### Instructions:
1. Run all cells in this notebook.
2. The exported files will be created in the current directory.

```python
import sqlite3
import pandas as pd
import json
import os

# Path to the database file
db_path = "../database/faculty.db"

if not os.path.exists(db_path):
    print(f"Error: Database not found at {db_path}")
else:
    # Connect to the database
    conn = sqlite3.connect(db_path)
    
    # Fetch all 109 faculty records
    query = "SELECT * FROM faculty"
    df = pd.read_sql_query(query, conn)
    
    # Close connection
    conn.close()
    
    print(f"Successfully loaded {len(df)} records from the database.")
    
    # Export to CSV
    csv_filename = "faculty_data_export.csv"
    df.to_csv(csv_filename, index=False)
    print(f"CSV File Created: {csv_filename}")
    
    # Export to JSON
    json_filename = "faculty_data_export.json"
    df.to_json(json_filename, orient="records", indent=4)
    print(f"JSON File Created: {json_filename}")
    
    print("\nDone! You can now use these files for your analysis.")
```
